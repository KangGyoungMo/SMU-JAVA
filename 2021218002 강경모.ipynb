```python
import gdown

gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
# prompt: # Load a CSV file named 'ns_book7.csv' and show the first 5 rows.

import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
# prompt: Scatter plot for points (1,1), (2,2), (3,3), and (4,4)

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
# prompt: # Create a scatter plot for '번호' on the x-axis and '대출건수' on the y-axis, then display it.

import matplotlib.pyplot as plt

# Assuming 'ns_book7' DataFrame is already loaded as shown in the provided code.
# If not, load it first:
# import pandas as pd
# ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)


plt.scatter(ns_book7['번호'], ns_book7['대출건수'])
plt.show()
```

```python
# prompt:  Scatter plot of '도서권수' against '대출건수' columns in ns_book7.

plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'])
plt.xlabel('도서권수')
plt.ylabel('대출건수')
plt.title('Scatter Plot of 도서권수 vs 대출건수')
plt.show()
```

```python
# prompt: # Create a scatter plot of '도서권수' (Book Count) vs '대출건수' (Borrow Count) with 10% opacity, and display the plot.

import matplotlib.pyplot as plt

# Assuming 'ns_book7' DataFrame is already loaded.
# If not, load it first:
# import pandas as pd
# ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)

plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

```python
# prompt: # Calculate the average number of borrows per book by dividing '대출건수' by '도서권수' columns in 'ns_book7' dataset
# # Create a scatter plot of average borrows vs. total borrows with transparency to manage data overlap
# # Display the scatter plot

# Calculate the average number of borrows per book
ns_book7['평균대출'] = ns_book7['대출건수'] / ns_book7['도서권수']

# Create the scatter plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.scatter(ns_book7['평균대출'], ns_book7['대출건수'], alpha=0.5) # Use alpha for transparency
plt.xlabel('Average Borrows per Book')
plt.ylabel('Total Borrows')
plt.title('Scatter Plot of Average Borrows vs. Total Borrows')
plt.grid(True) # Add a grid for better readability
plt.show()
```
